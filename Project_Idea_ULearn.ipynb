```python
import random

class MathAptitudeTest:
    def __init__(self, num_students=5, num_questions=5):
        self.num_students = num_students
        self.num_questions = num_questions
        self.students_data = []

    def generate_questions(self):
        questions = []
        for _ in range(self.num_questions):
            operand1 = random.randint(1, 10)
            operand2 = random.randint(1, 10)
            operator = random.choice(['+', '-', '*'])
            question = f"{operand1} {operator} {operand2}"
            questions.append((question, eval(question)))
        return questions

    def conduct_test(self):
        questions = self.generate_questions()

        for student_id in range(1, self.num_students + 1):
            print(f"\nStudent {student_id}'s Math Aptitude Test:")
            score = 0

            for i, (question, correct_answer) in enumerate(questions, 1):
                user_answer = input(f"Q{i}: {question} = ")
                try:
                    user_answer = int(user_answer)
                except ValueError:
                    print("Invalid input. Please enter a number.")
                    user_answer = 0

                if user_answer == correct_answer:
                    print("Correct!")
                    score += 1
                else:
                    print(f"Wrong! The correct answer is {correct_answer}.")

            self.students_data.append({"Student_ID": student_id, "Score": score})

    def provide_guidance(self):
        print("\nGuidance based on scores:")
        for student_data in self.students_data:
            student_id = student_data["Student_ID"]
            score = student_data["Score"]

            print(f"Student {student_id}: Scored {score} out of {self.num_questions} questions.")

            if score >= self.num_questions * 0.8:
                print("Excellent performance! Keep it up.")
            elif score >= self.num_questions * 0.6:
                print("Good performance. There is room for improvement.")
            else:
                print("It's important to focus on areas where improvement is needed.")

if __name__ == "__main__":
    math_test = MathAptitudeTest()
    math_test.conduct_test()
    math_test.provide_guidance()
9
```

```python
import random

class MathAptitudeTest:
    def __init__(self, num_students=2, num_questions=5):
        self.num_students = num_students
        self.num_questions = num_questions
        self.students_data = []

    def generate_questions(self):
        questions = []
        for _ in range(self.num_questions):
            operand1 = random.randint(1, 10)
            operand2 = random.randint(1, 10)
            operator = random.choice(['+', '-', '*'])
            question = f"{operand1} {operator} {operand2}"
            questions.append((question, eval(question), operator))
        return questions

    def provide_guidance_for_question(self, operator, operand1, operand2, correct_answer):
        if operator == '+':
            return f"For addition problems, {operand1} + {operand2} is {correct_answer}."
        elif operator == '-':
            return f"For subtraction problems, {operand1} - {operand2} is {correct_answer}."
        elif operator == '*':
            return f"For multiplication problems, {operand1} * {operand2} is {correct_answer}."

    def conduct_test(self):
        questions = self.generate_questions()

        for student_id in range(1, self.num_students + 1):
            print(f"\nStudent {student_id}'s Math Aptitude Test:")
            score = 0

            for i, (question, correct_answer, operator) in enumerate(questions, 1):
                user_answer = input(f"Q{i}: {question} = ")
                try:
                    user_answer = int(user_answer)
                except ValueError:
                    print("Invalid input. Please enter a number.")
                    user_answer = 0

                if user_answer == correct_answer:
                    print("Correct!")
                    score += 1
                else:
                    print(f"Wrong! {self.provide_guidance_for_question(operator, question.split()[0], question.split()[2], correct_answer)}")

            self.students_data.append({"Student_ID": student_id, "Score": score})

    def provide_guidance(self):
        print("\nGuidance based on scores:")
        for student_data in self.students_data:
            student_id = student_data["Student_ID"]
            score = student_data["Score"]

            print(f"Student {student_id}: Scored {score} out of {self.num_questions} questions.")

            if score >= self.num_questions * 0.8:
                print("Excellent performance! Keep it up.")
            elif score >= self.num_questions * 0.6:
                print("Good performance. There is room for improvement.")
            else:
                print("It's important to focus on areas where improvement is needed.")

if __name__ == "__main__":
    math_test = MathAptitudeTest()
    math_test.conduct_test()
    math_test.provide_guidance()
```
